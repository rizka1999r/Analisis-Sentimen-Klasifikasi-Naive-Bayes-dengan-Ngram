```python
#memanggil senticnet untuk proses pemberian score perkata
label_csv = pd.read_csv('/content/drive/MyDrive/Salinan senticnet.tsv', sep='\t')
label_df = pd.DataFrame(label_csv)
lexicon_dict = dict(zip(label_df['CONCEPT'], label_df['POLARITY INTENSITY']))
```

```python
lexicon_dict
```

```python
def sentimen_lexicon(t): #function untuk pemberian score 
  score = []
  polarity = ''
  for w in t.split():
    if w in lexicon_dict: 
      score.append(float(lexicon_dict.get(w).replace(',', '.')))

#Jika nilai score lebih dari sama dengan 1 polarity score Positive , Jika nilai Score kurang dari sama dengan 1 , jika tidak dari dua kondisi diatas score polarity neutral
  if sum(score) >= 1:
    polarity = 'positive'
  elif sum(score) <= -1:
    polarity = 'negative'
  # else:
  #   polarity = 'neutral'

    # print(t)
  # print("score={}, sum={}".format(score, sum(score)))
  return polarity, sum(score)
```

```python
#Pemanggilan function Lexicond dan pemberian polarity dan polarity_score
label_result = tweets['text_preprocessed'].apply(sentimen_lexicon)
label_result = list(zip(*label_result))
tweets['polarity'] = label_result[0]
tweets['polarity_score'] = label_result[1]
# tweets = tweets[tweets.polarity != 'neutral']
tweets = tweets.reset_index()
```

```python
tweets = tweets.drop('index', axis=1)
```
